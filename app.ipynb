```python
import pandas as pd
import matplotlib.pyplot as plt
```

Ler a base de dados e organiliza-la

```python
dados = pd.read_csv("fatura.csv")
dados
```

Iniciar a analise de dados da fatura

Calcular valor total da fatura

```python
valor_total_cartao = dados['ValorCompra'].sum()
print(f"O valor total gasto no cartão foi de R${valor_total_cartao:.2f}")
```

Valor gastor por titular

```python
valor_total_titular = dados.groupby('Nome')['ValorCompra'].sum().reset_index().sort_values(by='ValorCompra', ascending=False)
valor_total_titular
```

Custo gasto por categoria de compras

```python
valor_categoria = dados.groupby('categoria')['ValorCompra'].sum().reset_index().sort_values(by='ValorCompra', ascending=False)
valor_categoria
```

Dias de maiores consumos

```python
consumo_dias = dados.groupby('data_compra')['ValorCompra'].sum().reset_index().sort_values(by='data_compra', ascending=True)
consumo_dias
```

gastos por titular

```python
plt.figure(figsize=(10, 5))
plt.bar(valor_total_titular['Nome'], valor_total_titular['ValorCompra'], color='lightblue')
plt.title('Valor total gasto por titular')
plt.xlabel('Titular') 
plt.ylabel('Valor total gasto (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
```

gastos por categoria

```python
plt.figure(figsize=(10, 5))
plt.barh(valor_categoria['categoria'], valor_categoria['ValorCompra'], color='lightblue')
plt.title('Valor total gasto por categoria')
plt.xlabel('Valor gasto por categoria (R$)')
plt.ylabel('Categoria')
```

gastos por titular

```python
plt.figure(figsize=(5, 5))
plt.pie(valor_total_titular['ValorCompra'], labels=valor_total_titular['Nome'], autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Proporção de gastos por titular')
```
